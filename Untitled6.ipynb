```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Sample Dataset (10 entries)
# Features: [Average Frequency (Hz), Volume (dB)]
X = np.array([
    [1200, 50],
    [1350, 53],
    [1400, 55],
    [1600, 60],
    [1800, 63],
    [2000, 66],
    [2100, 70],
    [2300, 74],
    [2500, 76],
    [2700, 78]
])

plt.figure(figsize=(8,5))
sns.scatterplot(x=X[:, 0], y=X[:, 1], color='blue', s=100)
plt.title('Average Frequency (Hz) vs Volume (dB)')
plt.xlabel('Average Frequency (Hz)')
plt.ylabel('Volume (dB)')
plt.grid(True)
plt.show()

# Target: Fault Score (0–10 scale)
y = np.array([1.0, 2.0, 2.5, 4.0, 5.0, 6.5, 7.0, 8.0, 9.0, 9.8])

# Step 5: Define variables
X_features = X  # Feature matrix
y_target = y        # Target variable

# Step 6: Train Linear Regression model
model = LinearRegression()
model.fit(X_features, y_target)

# Step 7: Print intercept and coefficient
print(f"Intercept (b0): {model.intercept_}")
print(f"Coefficient (b1): {model.coef_}")

# Step 4: Make predictions
y_pred = model.predict(X_features)

# Step 5: Display predictions
print("\nPredicted Fault Scores:")
for i in range(len(X_features)):
    print(f"Input: Freq={X_features[i][0]}, Vol={X_features[i][1]} → Predicted: {round(y_pred[i], 2)} | Actual: {y_target[i]}")

# Step 6: Evaluation metrics
mse = mean_squared_error(y_target, y_pred)
r2 = r2_score(y_target, y_pred)
print("\nEvaluation Metrics:")
print("Mean Squared Error (MSE):", round(mse, 4))
print("R² Score:", round(r2, 4))

# Step 7: Plot regression (for 2D view, using frequency only)
freq = X_features[:, 0].reshape(-1, 1)
model_1d = LinearRegression()
model_1d.fit(freq, y_target)
y_pred_1d = model_1d.predict(freq)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(freq, y_target, color='blue', label='Actual')
plt.plot(freq, y_pred_1d, color='red', label='Regression Line')
plt.xlabel("Average Frequency (Hz)")
plt.ylabel("Fault Score")
plt.title("Regression Line (Freq vs Fault Score)")
plt.legend()
plt.grid(True)
```

```python
residuals = y - y_pred

plt.figure(figsize=(8,5))
sns.residplot(x=X[:, 0], y=residuals, lowess=True, color='green')
plt.title('Residual Plot')
plt.xlabel('Average Frequency (Hz)')
plt.ylabel('Residuals')
plt.axhline(0, color='red', linestyle='--')
plt.grid(True)
plt.show()
```
